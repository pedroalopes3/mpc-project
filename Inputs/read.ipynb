```python
input="Input-P2-0.txt"

f1 = open("2.csv", "w") 
```

```python
with open(input, "r") as f:
  i=0
  st=""
  for x in f:
    x=x.strip()
    if i in [0,1,2]:
        st += str(x)+","
        i+=1
    else:
        st += str(x)+"\n"
        i=0
        f1.write(st)
        st =""
	
f1.close() 
```
